Data Insights and Business Question Answer

#Questions

1) Find top 10 highest reveue generating products
2) Find top 5 highest selling products in each region
3) Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023
4) For each category which month had highest sales
5) Which sub category had highest growth by profit in 2023 compare to 2022

```python
df_order.head()
```

First Answer

```python
prod_rev = df_order.groupby(df_order['product_id'])['sales_price'].sum()
```

```python
prod_rev.sort_values(ascending = False).head(10)
```

Second Answer

```python
cte = df_order.groupby(['product_id', 'region']).agg(selling_prod_region=('sales_price', 'sum')).reset_index()
```

```python
cte['rn'] = cte.groupby('region')['selling_prod_region'].transform(lambda x: x.rank(method='min', ascending=False))
```

```python
result_df = cte[cte['rn'] <= 5]
```

```python
result_df.sort_values(by=['region', 'rn'], ascending=[True, True])
```

Third Answer

```python
df_order['month'] = df_order['order_date'].dt.month
```

```python
df_order['year'] = df_order['order_date'].dt.year
```

```python
res_df1 = df_order[df_order['year'] == 2022]
```

```python
res_df2 = df_order[df_order['year'] == 2023]
```

```python
res_df1 = res_df1.groupby(['month']).agg(sales_2022 = ('sales_price', 'sum')).reset_index()
```

```python
res_df2 = res_df2.groupby(['month']).agg(sales_2023 = ('sales_price', 'sum')).reset_index()
```

```python
result = pd.merge(res_df1, res_df2, on = 'month', how = 'inner')
```

```python
result
```

Fourth Answer

```python
df_order['month'] = df_order['order_date'].dt.month
```

```python
df_order['year'] = df_order['order_date'].dt.year
```

```python
condition = df_order['month'] <= 9
df_order['mon_year'] = df_order['year'].astype(str) + condition.map({True: '0', False: ''}) + df_order['month'].astype(str)
```

```python
cte = df_order.groupby(['category','mon_year']).agg(sales = ('sales_price', 'sum')).reset_index()
```

```python
cte = cte.groupby(['category']).agg(max_sales = ('sales', 'max')).reset_index()
```

```python
cte
```

Fifth Answer

```python
df_order['year'] = df_order['order_date'].dt.year
```

```python
cte1 = df_order[df_order['year'] == 2022]
```

```python
cte2 = df_order[df_order['year'] == 2023]
```

```python
cte1 = cte1.groupby(['sub_category']).agg(sales2022 = ('sales_price', 'sum')).reset_index()
```

```python
cte2 = cte2.groupby(['sub_category']).agg(sales2023 = ('sales_price', 'sum')).reset_index()
```

```python
result = pd.merge(cte1, cte2, on = 'sub_category', how = 'inner')
```

```python
result['profit'] = result['sales2023'] - result['sales2022'] 
```

```python
final = result.sort_values(by = 'profit', ascending = False).head(1)
```
